> Method chaining is a programmatic style of invoking multiple method calls sequentially with each call performing an action on the same object and returning it.
>
> It eliminates the cognitive burden of naming variables at each intermediate step. Fluent Interface, a method of creating object-oriented API relies on method cascading (aka method chaining). This is akin to piping in Unix systems.

```python
import pandas as pd
```

```python
data = pd.read_csv("titanic.csv")
```

```python
def split_name_series(string):
    first_name, second_name = string.split(", ")
    return pd.Series((first_name, second_name), index=["firstName", "secondName"])
```

```python
split_name_series(data.iloc[0, 3])
```

```python
def split_name(x_df):
    def split_name_series(string):
        first_name, second_name = string.split(", ")
        return pd.Series((first_name, second_name), index=["firstName", "secondName"])

    res = x_df["Name"].apply(split_name_series)
    x_df[res.columns] = res
    return x_df
```

```python
def split_name2(x_df):
    res = x_df.loc[:, "Name"].str.split(", ", n=1, expand=True)
    res.rename(columns={0: "first_name", 1: "second_name"}, inplace=True)
    x_df[res.columns] = res
    return x_df
```

```python
data.pipe(split_name2)
```

```python
def substitute_sex(x_df):
    mapping = {"male": "M", "female": "F"}
    x_df["Sex"] = x_df["Sex"].map(mapping)
    return x_df
```

```python
data.pipe(split_name2).pipe(substitute_sex)
```
